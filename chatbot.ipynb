IMPORTING LIB:

```python
import datetime
import random
import zoneinfo
```

WELCOME MSG

```python
def welcome_msg():
    print("Hello Dear! I am Rule Based Chatbot. Type your message to start chatting")
    print("You can talk to me about greetings, time, date or just chat or play game or ask me to do some calculation.")
    print("Type exit to exit")
```

GAME:

```python
def number_game():
  print("Game Started! Enter a number from 1 to 100")
  print("Enter Bye to exit game")
  number = random.randint(1, 100)
  while True:
    gamer_input = input("YOU: ")
    if gamer_input.isdigit():
      guess = int(gamer_input)
      if guess<number:
        print("Too Low")
      elif guess>number:
        print("Too High")
    elif guess == number:
      print("Congratulations! You Won")
      break
    elif gamer_input == "bye":
       break
    else:
       print("Invalid Input")
```

CALCULATOR:

```python
def calculator():
  print("Enter numbers to perform calculations")
  print("Enter 1 for Addition")
  print("Enter 2 for Subtraction")
  print("Enter 3 for Multiplication")
  print("Enter 4 for Division")
  print("Enter 5 to exit")
  while True:
    choice = int(input("Enter your choice: "))
    if choice == 5:
      break
    num1 = float(input("Enter First Number: "))
    num2 = float(input("Enter Second Number: "))
    if choice == 1:
     print("The sum is",num1+num2)
    elif choice == 2:
     print("The difference is",num1-num2)
    elif choice == 3:
     print("The Multiplication is",num1*num2)
    elif choice == 4:
     if num2!=0:
        print("The division is",num1/num2)
    else:
        print("Division by Zero is not allowed.")
  else:
    print("Invalid Input")
```

RULES:

```python
def chatbot_response(user_input):
  user_input = user_input.lower()
  if "hello" in user_input or "hi" in user_input or "hey" in user_input:
    return "Hii, How can I help You Today!"
  elif "who are you" in user_input or "what do you do" in user_input:
      return "I am a rule based chatbot, I can talk with you and play a random number choosing game with you!"
  elif "name" in user_input:
      return "My name is BuddyBot"
  elif "how are you" in user_input:
      return "I am doing great, thanks for asking!"
  elif "who created you" in user_input:
      return "I was created by Nancy"
  elif "How can you help me" in user_input or "what can you do" in user_input:
      return "I can chat with you, Play game with you, Perform some basic calculation for you or Can give you some study tips to manage stress during exams."
  elif "calculation" in user_input or "addition" in user_input or "subtract" in user_input or "multiply" in user_input or "divide" in user_input:
      calculator()
      return "Calculation Over"
  elif "study tips" in user_input:
      return "Study for 25 minutes then take 5 minute break will help ypu learn better"
      return "Revise your notes daily"
  elif "exam" in user_input:
      return "Exam Tip:Make a realistic timetable and focus more on weak subjects"
  elif "game" in user_input :
    number_game()
    return "Game Over"
  elif"time" in user_input:
    tz = zoneinfo.ZoneInfo("Asia/Kolkata")
    current_time = datetime.datetime.now(tz).strftime("%H:%M:%S")
    return f"The current time is {current_time}."
  elif "date" in user_input:
        current_date = datetime.datetime.now().strftime("%Y-%m-%d")
        return f"Today's date is {current_date}."
  elif "bye" in user_input or "goodbye" in user_input or "exit" in user_input:
    return "exit"
  else:
    return "I didn't quite understand that. Can you rephrase?"

```

MAIN CODE:

```python
def chatbot():
    welcome_msg()
    while True:
      user_input = input("YOU: ")
      response = chatbot_response(user_input)
      if "exit" in response:
        print("Goodbye! Have a wonderful day.")
        break
      else:
        print("Chatbot:", response)
chatbot()

```
